### Часть 1. Задача 1.

Произошло убийство. На месте убийства найдена кровь, которая явно (будем
считать, что с вероятностью 1) принадлежит настоящему убийце. Кровь
принадлежит редкой группе, которая присутствует всего лишь у 1% населения. И так
получилось, что у подсудимого как раз эта редкая группа крови!
Первым выходит прокурор и говорит: “Шанс, что у подсудимого была бы именно
такая группа крови, если бы он был невиновен -- всего 1%; значит, с вероятностью
99% он виновен, предлагаю осудить”. В чём не прав прокурор? Укажите, какие
вероятности он оценил, а какие должен был оценить.

Обозначим:
- $A$ - {подозреваемый виновен}
- $B$ - {наличие редкой группы крови}
- $P(A)$ - вероятность того что подозреваемый виновен
- $P(\overline{A})$ - вероятность того что подозреваемый невиновен
- $P(B) = 0.01$ - вероятность редкой группы крови
- $P(\overline{B}) = 1 - P(B) = 0.99$ - вероятность иметь не редкую группу крови

По условию задачи убийца имел редкую группу крови, это условие возможно записать через условную вероятность:
- $P(B|A) = 1$ 


Прокурор оценил $P(\overline{B}) = 0.99$ и $P(B) = 0.01$, а должен был оценить условную вероятность $P(A|B)$, то есть вероятность того, что подозреваемый виновинев при условии редкой группы крови:

$$
P\left( A \vert B   \right) = \dfrac{P\left(A\right) P\left( B \vert A   \right)}{P \left( B \right)} = 
\dfrac{P\left(A\right) \cdot 1}{0.01} = 100 P\left(A\right)
$$

### Часть 1. Задача 3.

Вернёмся к тесту на страшную болезнь из первой лекции. Напомню, что по
предположениям задачи болеет 1% населения, а тест имеет вероятность ошибки
5% в обе стороны. Мы видели, что в результате человек с положительным тестом
имеет апостериорную вероятность оказаться больным всего лишь около 16%.
Врачи решили, что терпеть это никак нельзя; они готовы провести срочное
исследование и улучшить качество теста. Но сил и денег хватит только для того,
чтобы сократить один тип ошибок, т.е. уменьшить или только число false positives
(когда тест выдаёт положительный результат у здорового человека), или только
число false negatives (когда тест выдаёт отрицательный результат у больного).
Помогите медикам: постройте зависимости апостериорной вероятности болезни
после положительного и отрицательного теста от обоих видов ошибок (например, в
виде графиков), сделайте выводы и дайте рекомендации врачам -- на чём лучше
сосредоточиться?

Обозначим:

- $A$ - {человек болен}
- $B$ - {тест положителен}
- $FN$ - величина (доля) ошибки false negatives (тест выдаёт отрицательный результат у больного)
- $FP$ - величина (доля) ошибки false positives (тест выдаёт положительный результат у здорового человека)
- $P(A) = 0.01$ - вероятность того, что человек болен
- $P(\overline{A}) = 0.99$ - вероятность того, что человек здоров
- $P(B|A) = 1 - FN$ - вероятность положительного теста при условии, что человек болен
- $P(B|\overline{A}) = FP$ - вероятность положительного теста при условии, что человек здоров
- $P(\overline{B}|A) = FN$ - вероятность отрицательного теста при условии, что человек болен
- $P(\overline{B}|\overline{A}) = 1 - FP$ - вероятность отрицательного теста при условии, что человек здоров

Тогда:

- вероятность того, что человек болен при положительном тесте

$$
P(A|B) = \dfrac{P(A) P(B|A)}{P(A) P(B|A) + P(\overline{A})P(B|\overline{A})} = 
\dfrac{P(A) (1 - FN)}{P(A) (1 - FN) + P(\overline{A}) FP}
$$

- вероятность того, что человек болен при отрицательном тесте

$$
P(A|\overline{B}) = \dfrac{P(A) P(\overline{B}|A)}{P(A) P(\overline{B}|A) + P(\overline{A})P(\overline{B}|\overline{A})} = 
\dfrac{P(A) FN}{P(A) FN + P(\overline{A}) (1 - FP)}
$$

```python
import numpy as np
from matplotlib import pyplot as plt
plt.rcParams.update({'font.size': 12})

def P_A_B(FN, FP, P_A = 0.01):
    return P_A * (1 - FN) / (P_A * (1 - FN) + (1 - P_A) * FP)

def P_A_notB(FN, FP, P_A = 0.01):
    return P_A * FN / (P_A * FN + (1 - P_A) * (1 - FP))
```

```python
FN_range = np.arange(0, 0.05, 0.0005)
FP_range = FN_range

fig = plt.figure()
axes = fig.add_axes([0, 0, 1, 1])
axes.plot(FN_range, P_A_B(FN=FN_range, FP=0.05), 'green', label='Тест положительный')
axes.plot(FN_range, P_A_notB(FN=FN_range, FP=0.05), 'blue', label='Тест отрицательный')
axes.set_ylim((0,1))
axes.set_title('Вероятность того, что человек болен при FN от 0 до 0.05')
axes.set_ylabel('Bероятность того, что человек болен')
axes.set_xlabel('FN')
axes.grid()
axes.legend(loc = 1);

ax = fig.add_subplot()
ax.set_position([1.2, 0, 1, 1])
ax.plot(FN_range, P_A_B(FN=0.05, FP=FP_range), 'green', label='Тест положительный')
ax.plot(FN_range, P_A_notB(FN=0.05, FP=FP_range), 'blue', label='Тест отрицательный')
ax.set_ylim((0,1))
ax.set_title('Вероятность того, что человек болен при FP от 0 до 0.05')
ax.set_ylabel('Bероятность того, что человек болен')
ax.set_xlabel('FP')
ax.grid()
ax.legend(loc = 1);
plt.show();
```

### Вывод
- Построенные зависимости апостериорной вероятности болезни после положительного и отрицательного теста для обоих видов ошибок демострируют, что наибольший эффект в увелечении точности диагности после теста возможно получить при уменьшении ошибки FP (положительный результат у здорового человека)

#### Часть 2. Задача 2.b

```python
from scipy.stats import multivariate_normal

def bayesian_update_for_array(mu, sigma, x, y, sigma_noise=.25):
    x_matrix = np.hstack((np.ones_like(x), x))
    sigma_n = np.linalg.inv(np.linalg.inv(sigma)+ (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x_matrix), x_matrix))
    mu_n = np.matmul(sigma_n, np.matmul(np.linalg.inv(sigma), np.transpose(mu)) + (1 / (sigma_noise ** 2)) * np.matmul(np.transpose(x_matrix), np.array(y)))
    return mu_n, sigma_n
```

```python
noise = np.std(np.log(data_train["total_cases"]) - np.log(data_train["y_pred"]))
cur_mu, cur_sigma = np.array([0, 0]), 3*np.array([[1, 0], [0, 1]])
cur_mu, cur_sigma = bayesian_update_for_array(cur_mu, cur_sigma, 
                                              data_train['day_number'].values.reshape(-1, 1), 
                                              np.log(data_train['total_cases'].values), 
                                              sigma_noise = noise)

print(f"Стандартное отклонение факта от модели (оценка шума): {noise}")
print(f"Оценка мат. ожидания: {cur_mu}")
print(f"Оценка стандартного отклонения:\n {cur_sigma}")
```

```python
N = 3000
X = np.linspace(-1, 2, N)
Y = np.linspace(-1, 2, N)
X, Y = np.meshgrid(X, Y)

pos = np.empty(X.shape + (2,))
pos[:, :, 0] = X
pos[:, :, 1] = Y

def myplot_heatmap(Z, xlim=(-1, 2), ylim=(-1, 2)):
    plt.xlim(xlim)
    plt.ylim(ylim)
    plt.pcolormesh(X, Y, Z, cmap=plt.cm.jet, shading='auto')

fig = plt.figure(figsize=(10, 10))

Z = multivariate_normal.pdf(pos, mean=cur_mu, cov=cur_sigma)
myplot_heatmap(Z, (1.5, 2), (0, 0.5))
```

#### Часть 2. Задача 3.a

Строится модель следующего вида:

$$
\large{y = \int_{-\infty}^{t} e^{ax^2 + bx + c} \,dx}
$$

В качестве аппроксимации воспользуемся приближенным выражением:

$$
\Delta y (x) \approx y \prime (x) \Delta x = e^{ax^2 + bx + c} \Delta x
$$

С учетом того, что шаг в задаче постоянный ($\Delta x = 1$) возможно модель искать в следующем виде

$$
\log(\Delta y (x)) = ax^2 + bx + c
$$

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression()
data_train["day_number_2"] = data_train["day_number"] ** 2

X = data_train[["day_number_2", "day_number"]]
data_train["delta_total_cases"] = data_train["total_cases"].diff()

# Вынужденные корретировки для операции логарифмирования
data_train["delta_total_cases"].fillna(1, inplace = True)
data_train["delta_total_cases"] = data_train["delta_total_cases"].apply(lambda x: x if x != 0 else 1)

y = np.log(data_train["delta_total_cases"])
lr.fit(X, y)
data_train["delta_y_pred"] = np.exp(lr.predict(X))
print(f"Параметры модели a = {lr.coef_[0]:.4f}, b = {lr.coef_[1]:.4f},  c = {lr.intercept_:.4f}")
data_train["y_pred2"] = data_train["delta_y_pred"].cumsum() + 3
```

```python
pl = data_train.plot(x = 'date', y = ['total_cases', 'y_pred2'], kind = 'bar', title="Число заболевших", 
                    figsize=(17,10), grid = True, label = ['Общее количество (факт)', 'Общее количество (модель)']);
pl.set_ylabel("Kоличество");
pl.set_xlabel("Дата");
```
